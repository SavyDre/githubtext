```python
from IPython.display import clear_output
clear_output()

board = {'1':' ','2':' ','3':' ','4':' ','5':' ','6':' ','7':' ','8':' ','9':' '}

boardkeys = []
for key in board:
    boardkeys.append(key)
    
def displayboard(board):
    
    print('| ' + board['1'] + ' | ' + board['2'] + ' | ' + board['3'] + ' |')
    print("+---+---+---+")
    print('| ' + board['4'] + ' | ' + board['4'] + ' | ' + board['6'] + ' |')
    print("+---+---+---+")
    print('| ' + board['7'] + ' | ' + board['8'] + ' | ' + board['9'] + ' |')
        
def choosemarker():
    
    marker = 'wrong'    
    while marker not in ['X','O']:
        marker = input("Player, please select your marker, it is either X or O: ").upper()
        if marker not in ['X','O']:
            print('Marker must be X or O, try again')
            continue
    if marker == 'X':
        return ('X','O')
    else:
        return ('O','X')
    
def who_goes_first():
    import random
    if random.randint(0,1) == 1:
        return 'Player 1'
    else:
        return 'Player 2'
    
def lets_play():
    clear_output()
    player1_marker,player2_marker = choosemarker()
    player = who_goes_first()
    count = 0
    
    for i in range(9):
        displayboard(board)
        print(player + ", please make your next move")
        move = input('Input a digit between 1 to 9: ')
        
        if board[move] == ' ':
            if player == 'Player 1':
                board[move] = player1_marker
                count = count + 1
                clear_output()
            else:
                board[move] = player2_marker
                count = count + 1
                clear_output()
                
        else:
            print("The space you are trying to fill is occupied, try again")
            continue
        
        if count >= 5:
            if board['1'] == board['2'] == board['3'] != ' ':
                displayboard(board)
                print('Congratulations ' + player + ', you have won the game')
                break
            elif board['4'] == board['5'] == board['6'] != ' ':
                displayboard(board)
                print('Congratulations ' + player + ', you have won the game')
                break
            elif board['7'] == board['8'] == board['9'] != ' ':
                displayboard(board)
                print('Congratulations ' + player + ', you have won the game')
                break
            elif board['1'] == board['4'] == board['7'] != ' ':
                displayboard(board)
                print('Congratulations ' + player + ', you have won the game')
                break
            elif board['2'] == board['5'] == board['8'] != ' ':
                displayboard(board)
                print('Congratulations ' + player + ', you have won the game')
                break
            elif board['3'] == board['6'] == board['9'] != ' ':
                displayboard(board)
                print('Congratulations ' + player + ', you have won the game')
                break
            elif board['1'] == board['5'] == board['9'] != ' ':
                displayboard(board)
                print('Congratulations ' + player + ', you have won the game')
                break
            elif board['3'] == board['5'] == board['7'] != ' ':
                displayboard(board)
                print('Congratulations ' + player + ', you have won the game')
                break
        if count == 9:
            print('The game is a tie')
            
        if player == 'Player 1':
            player = 'Player 2'
        else:
            player = 'Player 1'
            
        replay = 'wrong'
        while replay not in ['YES','NO']:
            replay = input("Do you want to play again? answer Yes or No: ").upper()
            if replay not in ['YES','NO']:
                print("To chose if you want to play again you need to select either yes or no")
            
        for key in boardkeys:
            board[key] == " "
            lets_play()

lets_play()    
```
